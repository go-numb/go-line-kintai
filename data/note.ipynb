```python
import pandas as pd
```

```python
# ファイル変数
path = ""
df = pd.read_csv(path)
```

```python
df.describe()
```

```python
## 当月集計
# 当月のデータを抽出
now = pd.Timestamp.now()
this_year = now.year
this_month = now.month
print(this_year, this_month)
df_this_month = df[(df["year"] == this_year) & (df["month"] == this_month)]
df_this_month.describe()
```

```python
## 前月集計
# 前月のデータを抽出
now = pd.Timestamp.now() - pd.DateOffset(months=1)
this_year = now.year
prev_month = now.month
print(this_year, this_month)
df_prev_month = df[(df["year"] == this_year) & (df["month"] == prev_month)]
df_prev_month.describe()
```
